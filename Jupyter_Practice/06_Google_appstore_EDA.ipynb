```python
# Convert the whole size column into bytes 
# Let's define a function

def convert_size(size):
    if isinstance(size, str):
        if 'k' in size:
            return float(size.replace('k', '')) * 1024
        elif 'M' in size:
            return float(size.replace('M', '')) * 1024 * 1024
        elif 'Varies with device' in size:
            return np.nan
    return size
```

```python
# Let's apply the function
df['Size'] = df['Size'].apply(convert_size)
```

```python
df['Size']
```

```python
# Rename the column
df.rename(columns = {'Size':'Size_in_bytes'}, inplace = True)
```

```python
df.head()
```

```python
df['Size_in_MB'] = df['Size_in_bytes'].apply(lambda x: x/(1024*1024))
```

```python
df['Size_in_KB'] = df['Size_in_bytes'].apply(lambda x: x/(1024))
```

# Install column 

1. Remove + sign
2. Remove ,
3. Convert the column into int


```python
df['Installs'] = df['Installs'].apply(lambda x: x.replace('+', '') if '+' in str(x) else x)
```

```python
df['Installs'] = df['Installs'].apply(lambda x: x.replace(',', '') if ',' in str(x) else x)
```

```python
df['Installs'] = df['Installs'].apply(lambda x : int(x))
```

# Price Column 

```python
df['Price'].value_counts()
```

```python
# Let's see how many value have & sign
df['Price'].loc[df['Price'].str.contains('\$')].value_counts().sum()
```

```python
df['Price'].loc[df['Price'].str.contains('\$')].value_counts() & (~df['Price'].loc[df['Price'].str.contains('\$')].value_counts().sum())
```

```python
df['Price'] = df['Price'].apply(lambda x: x.replace('$', '') if '$' in str(x) else x)
```

```python
df.describe()
```

```python
# Now we can convert into numberic value
df['Price'] = df['Price'].apply(lambda x : float(x))
```

```python
df['Price'].value_counts()
```

```python
# using f string print min, max, and average price of the app
print(f"Minimun price is {df['Price'].min()}")
print(f'Max price is {df["Price"].max()}')
print(f'Average price is {df["Price"].mean()}')
```
